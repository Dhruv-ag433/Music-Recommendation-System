```python
import pandas as pd
```

```python
df1 = pd.read_csv("../dataset/BollyWood-Songs.csv")
df2 = pd.read_csv("../dataset/Mix-Songs.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
new_column_names = {
    'artist_name': 'singer',
    'track_name': 'music_name',
    'release_date': 'release'}

df2.rename(columns=new_column_names, inplace=True)
df2.head()
```

```python
df1_columns = set(df1.columns)

df2_selected = df2[list(df1_columns.intersection(df2.columns))]

df_merged = pd.concat([df1, df2_selected], axis=0, ignore_index=True)

df_merged.head()
```

```python
df_merged.shape
```

```python
df_merged.isnull().sum()
```

```python
default_thumbnail = "../dataset/default.png"
df_merged['thumbnail'].fillna(default_thumbnail, inplace=True)
```

```python
df_merged.isnull().sum()
```

```python
df_merged.to_csv("../dataset/final.csv", index=False)
```
